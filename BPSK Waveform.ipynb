### Generating BPSK

```python
import numpy as np
import numpy as np
import scipy
import matplotlib.pylab as plt
samples= 500
unipolardata=np.array([1,0, 1, 1, 0])
x= unipolardata.size
t= np.linspace(0,x,samples)
y=100*x
t1=np.linspace(0,100*x,samples)

sinewave= np.sin(2*np.pi*10*t1)

polardata= 2*unipolardata-1
unipolarwave= np.repeat(unipolardata,samples/x )
polarwave= np.repeat(polardata,samples/x )

plt.plot(t1,sinewave)
plt.xlabel('time (s)')
plt.ylabel('amplitude(V)')
plt.grid()
plt.show()
plt.plot(polarwave)
plt.xlabel('time (s)')
plt.ylabel('amplitude(V)')
plt.title('Polar NRZ')
plt.grid()
plt.show()

BPSKout= polarwave*sinewave
plt.plot(t1,BPSKout)
plt.xlabel('time (s)')
plt.ylabel('amplitude(V)')
plt.title('BPSKout')
plt.grid()
plt.show()

```
