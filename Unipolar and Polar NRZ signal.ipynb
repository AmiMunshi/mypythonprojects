## Generating Unipolar and Polar NRZ signal

```python
import numpy as np
import numpy as np
import scipy
import matplotlib.pylab as plt

unipolardata=np.array([1,0, 1, 1, 0])
x= unipolardata.size
polardata= 2*unipolardata-1
samples= 500
t= np.linspace(0,5,samples)
unipolarwave= np.repeat(unipolardata,samples/x )
polarwave= np.repeat(polardata,samples/x )
plt.plot(unipolarwave)
plt.xlabel('time (s)')
plt.ylabel('amplitude(V)')
plt.title('Unipolar NRZ')
plt.grid()
plt.show()
plt.plot(polarwave)
plt.xlabel('time (s)')
plt.ylabel('amplitude(V)')
plt.title('Polar NRZ')
plt.grid()
plt.show()
```
